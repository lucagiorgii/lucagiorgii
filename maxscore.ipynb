```python
import math
import pandas as pd
import numpy as np
```

```python
#load data from excel file using pip or conda install xlrd
#install pip 
#install conda
#install xlrd

df = pd.read_csv('heart_data1.csv', index_col=0)
df.head()


```

```python
#descrive data
df.describe()

```

```python
#check for null values
df.isnull().sum()
```

```python
#check for duplicates
df.duplicated().sum()

#drop duplicates
df.drop_duplicates(inplace=True)
df
```

```python
#fill numerical columns with median values: resting blood pressure, cholesterol, max heart rate, oldpeak,ca
df['resting blood pressure'] = df['resting blood pressure'].fillna(df['resting blood pressure'].median())
df['cholesterol'] = df['cholesterol'].fillna(df['cholesterol'].median())
df['max_heart_rate'] = df['max_heart_rate'].fillna(df['max_heart_rate'].median())
df['oldpeak'] = df['oldpeak'].fillna(df['oldpeak'].median())
df['ca'] = df['ca'].fillna(df['ca'].median())
df
```
